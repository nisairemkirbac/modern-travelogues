```python
import glob
from bs4 import BeautifulSoup
import re
import operator
import json
```

```python
#belgiantravelblogger
#make a dictionary with a list in it

btb_json_dict = {}
for filename in glob.glob ("#belgiantravelblogger/*.json"): 
        with open (filename) as jsonFile:
            data = json.loads (jsonFile.read ())
            print(data ["node"]["id"]) #make an empty list for every id. fill the dictionary with the keys (id in that case) and value (empty list).
            btb_json_dict [data ["node"]["id"]] = []
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["edge_media_to_caption"]["edges"][0]["node"]["text"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["shortcode"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["edge_liked_by"]["count"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["owner"]["id"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["accessibility_caption"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["iphone_struct"]["user"]["username"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["iphone_struct"]["user"]["full_name"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["iphone_struct"]["commerciality_status"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["iphone_struct"]["is_paid_partnership"])
            btb_json_dict [data ["node"]["id"]].append (data ["node"]["iphone_struct"]["comment_count"])
```

```python
print (btb_json_dict)
```
